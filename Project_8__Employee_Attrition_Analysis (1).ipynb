# Employee Attrition Analysis

## Problem Statement
XYZ company which was established a few years back is facing around a 15% attrition rate for a couple of years. And it's majorly affecting the company in many aspects. In order to understand why employees are leaving the company and reduce the attrition rate XYZ company has approached an HR analytics consultancy for analyzing the data they have. You are playing the HR analyst role in this project and building a dashboard which can help the organization in making data-driven decisions.

**Column Description**
1. EmployeeID (int64): Unique identifier for each employee.
2. Age (int64): Age of the employee.
3. Attrition (object): Indicates if the employee has left the company ('Yes') or not ('No').
4. BusinessTravel (object): Frequency of travel for business purposes.
5. Department (object): Department in which the employee works.
6. DistanceFromHome (int64): Distance of the employee's home from the workplace.
7. Education (int64): Education level of the employee.(1-below college,2-college,3-bachelor,4-Master,5-Doctor)
8. EducationField (object): Field of education of the employee.
9. EmployeeCount (int64): Number of employees (constant value of 1 for each record).
10. Gender (object): Gender of the employee.
11. JobLevel (int64): Level of the employee's job.
12. JobRole (object): Role of the employee within the company.
13. MaritalStatus (object): Marital status of the employee.
14. MonthlyIncome (int64): Monthly income of the employee.
15. NumCompaniesWorked (float64): Number of companies the employee has worked for.
16. Over18 (object): Indicator if the employee is over 18 (constant value of 'Yes').
17. PercentSalaryHike (int64): Percentage increase in salary.
18. StandardHours (int64): Standard hours worked (constant value of 80).
19. StockOptionLevel (int64): Stock option level of the employee.
20. TotalWorkingYears (float64): Total number of years the employee has worked.
21. TrainingTimesLastYear (int64): Number of training times last year.
22. YearsAtCompany (int64): Number of years the employee has been with the company.
23. YearsSinceLastPromotion (int64): Number of years since the employee's last promotion.
24. YearsWithCurrManager (int64): Number of years the employee has worked with their current manager.
25. EnvironmentSatisfaction (float64): Employee's satisfaction with the work environment.(1-Low, 2-Medium, 3-High,4-Very High)
26. JobSatisfaction (float64): Employee's job satisfaction.(1-Low, 2-Medium, 3-High,4-Very High)
27. WorkLifeBalance (float64): Employee's work-life balance.(1-Bad, 2-Good, 3-Better,4-Best)
28. JobInvolvement (int64): Employee's level of job involvement.(1-Low, 2-Medium, 3-High,4-Very High)
29. PerformanceRating (int64): Performance rating of the employee.(1-Low, 2-Good, 3-Excellent,4-Outstanding)

#### Attrition based on Age group

```python
def categorize_age(age):
    if age <= 20:
        return '<20'
    elif 20 < age <= 25:
        return '20-25'
    elif 25 <= age <= 30:
        return '25-30'
    elif 30 < age <= 35:
        return '30-35'
    elif 35 < age <= 40:
        return '35-40'
    elif 40 < age <= 45:
        return '40-45'
    elif 45 < age <= 50:
        return '45-50'
    else:
        return '>50'

data['AgeGroup'] = data['Age'].apply(categorize_age)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='AgeGroup', hue='Attrition', data=data, palette='Accent',order=data['AgeGroup'].value_counts().index)
plt.title('Attrition by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Observation:**

The employees of age group between 30-40 are more likely to leave the company.

#### Attrition based on the MonthlyIncome

```python
def salary(MonthlyIncome):
    if MonthlyIncome <= 20000:
        return '<20000'
    elif 20000 < MonthlyIncome<=40000:
        return '20000-40000'
    elif 40000 < MonthlyIncome<=70000:
        return '40000-70000'
    elif 70000 < MonthlyIncome<=100000:
        return '70000-100000'
    else:
        return '>100000'

data['Salary'] = data['MonthlyIncome'].apply(salary)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Salary', hue='Attrition', data=data, palette='husl',order=data['Salary'].value_counts().index)
plt.title('Attrition by MonthlyIncome')
plt.xlabel('MonthlyIncome')
plt.ylabel('Count')
plt.show()
```

**Observation:**

The employees who earns between 20K - 40K are more likely to leave the organization.

```python
sns.barplot(x=jobrole_wlb, y=jobrole_wlb.index, palette='magma')
plt.title('Average Work-Life Balance by Job Role')
plt.xlabel('Average Work-Life Balance')
plt.ylabel('Job Role')
plt.show()
```

```python
dept_wlb = data.groupby('Department')['WorkLifeBalance'].mean().sort_values()
```

```python
sns.barplot(x=dept_wlb, y=dept_wlb.index, palette='viridis')
plt.title('Average Work-Life Balance by Department')
plt.xlabel('Average Work-Life Balance')
plt.ylabel('Department')
plt.show()
```

```python
travel_wlb = data.groupby('BusinessTravel')['WorkLifeBalance'].mean().sort_values()
```

```python
sns.barplot(x=travel_wlb, y=travel_wlb.index, palette='coolwarm')
plt.title('Average Work-Life Balance by Business Travel Frequency')
plt.xlabel('Average Work-Life Balance')
plt.ylabel('Business Travel Frequency')
plt.show()
```

**Observations:**
* Managers has a much more average work life balance compared to other job roles.
* Sales department has a better work life balance than other departments.
* Worklife balance remains constant based on Travel.
